```python
import re
import time
from googletrans import Translator
import goslate
j=service_urls=[ 'https://translate.google.com/', 'https://translate.google.co.kr',]
```

```python
translator = Translator(timeout=10000,proxies='https://localhost/1290',user_agent="google-xrawler")
#translator=goslate.Goslate()
time.sleep(.10)
translator.translate('Belanjanya jadi banyak,','en').text
```

```python
with open ("episode 10.ass",'r+') as l:
    try:
        lines=l.readlines()
        print(len(lines))
        for i,line in enumerate(lines):
        
            time.sleep(.10)
            m = re.match( r'(.*),,0,0,0,,(.*?) .*', line, re.M|re.I)
            if m:
                print (i)
                lines[i]=translator.translate(m.group(),src='id',dest='en').text.encode('cp850','replace').decode('cp850')
    except:
        with open ("new10.ass",'w+') as new:
            new.writelines(lines)
            print('progress saved')
with open ("new10.ass",'w+') as new:
    new.writelines(lines)
    print('progress saved')
```
